### multiple plots

#### draw another graph on another layout(=figure)

```python
plt.plot(x,y1)
plt.figure()
plt.plot(x,y2)
plt.show()
```

```python
# close the figure
plt.close()
```

clear all graph on the figure
 - plt.clf()
 - plt.figure(N) : select(or indicate)d figure no.N

```python
# generate data
x = np.arange(-5,5,0.1)
y1 = x**2 - 2
y2 = 20 * np.cos(x)**2

plt.figure(1)    #generate space no.1
plt.plot(x, y1)  #draw graph_01 on no.1 figure

plt.figure(2)    #generate space no.2
plt.plot(x,y2)   #draw graph_02 on no.2 figure

plt.figure(1)    #indicate space no.1
plt.plot(x,y2)   #draw one more graph(graph_02) on no.1 figure

plt.figure(2)    #indicate space no.2
plt.clf()        #clear all graphs on no.2 figure
plt.plot(x,y1)   #draw graph_01 on no.2 figure

plt.show()
```

### subplots

```python
import numpy as np

x = np.arange(0,10,0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2 
y4 = 10*np.exp(-x) + 1 

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
import numpy as np

x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x, y1, '>--r',
         x, y2, 's-g',
         x, y3, 'd:b',
         x, y4, '-.Xc')

plt.legend(['data1', 'data2', 'data3', 'data4'])   # add legend

plt.show()
```

```python
import numpy as np

x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x, y1, '>--r',
         x, y2, 's-g',
         x, y3, 'd:b',
         x, y4, '-.Xc')

plt.legend(['data1', 'data2', 'data3', 'data4'],
           loc = 'lower right')   # set position for legend

plt.show()
```

```python
import matplotlib.pyplot as plt

height = [165, 177, 160, 180, 185, 155, 172]  
weight = [62, 67, 55, 74, 90, 43, 64]

plt.scatter(height, weight, s=300, c='r')   # size of marker : 300 / color of marker : red
plt.xlabel('Height(m)')
plt.ylabel('Weight(Kg)')
plt.title('Height&Weight')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

city = ['Seoul', 'Incheon', 'Daejeon', 'Daegu', 'Ulsan', 'Busan', 'Gwangju']

lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den)*0.2   #size of marker
colors = ['r','g','b','c','m','k','y']   #color of marker

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('population density 2017')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name)

plt.show()
```

## Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

member_IDs = ['mA_01', 'm_02', 'm_03', 'm_04']   # ID
before_ex = [27, 35, 40, 33]   # sit-up count before exercise
after_ex = [30, 38, 42, 37]   # sit-up count after exercise

n_data = len(member_IDs)    # 4 people -> 4 IDs
index = np.arange(n_data)   # generate array (0,1,2,3)
plt.bar(index, before_ex)   # bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

member_IDs = ['mA_01', 'm_02', 'm_03', 'm_04']  
before_ex = [27, 35, 40, 33]   
after_ex = [30, 38, 42, 37]  

n_data = len(member_IDs)    
index = np.arange(n_data)   
plt.bar(index, before_ex, tick_label=member_IDs)   # tick_label option : 'member_IDs' list
plt.title('give name on x-axis tick_label')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

member_IDs = ['mA_01', 'm_02', 'm_03', 'm_04']  
before_ex = [27, 35, 40, 33]   
after_ex = [30, 38, 42, 37]  

n_data = len(member_IDs)    
index = np.arange(n_data)   
colors = ['r','g','b','m']   # generate list for colors

plt.bar(index, before_ex, tick_label=member_IDs, color=colors)   # set color option
plt.title('set colors for bars')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

member_IDs = ['mA_01', 'm_02', 'm_03', 'm_04']  
before_ex = [27, 35, 40, 33]   
after_ex = [30, 38, 42, 37]  

n_data = len(member_IDs)    
index = np.arange(n_data)   
colors = ['r','g','b','m']   # generate list for colors

plt.bar(index, before_ex, tick_label=member_IDs, color=colors, width=0.5)   # set width of the bar
plt.title('set colors for bars')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

member_IDs = ['mA_01', 'm_02', 'm_03', 'm_04']  
before_ex = [27, 35, 40, 33]   
after_ex = [30, 38, 42, 37]  

n_data = len(member_IDs)    
index = np.arange(n_data)   
colors = ['r','g','b','m']   

plt.barh(index, before_ex, tick_label=member_IDs, color=colors)   # use 'barh()' function
plt.title('set colors for bars')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

member_IDs = ['mA_01', 'm_02', 'm_03', 'm_04']  
before_ex = [27, 35, 40, 33]   
after_ex = [30, 38, 42, 37]  

n_data = len(member_IDs)    
index = np.arange(n_data)   
barWidth = 0.4

plt.bar(index, before_ex, color='c', align='edge', width=barWidth, label='before')
plt.bar(index + barWidth, after_ex, color='m', align='edge', width=barWidth, label='after')

plt.xticks(index+barWidth, member_IDs)
plt.legend()
plt.xlabel('ID')
plt.ylabel('sit-up count')
plt.title('comparison before and after change')

plt.show()
```
